```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import matplotlib
```

```python
df = pd.read_csv('Natural_Gas_Pricing_DataSet.csv')
df['Date'] = pd.to_datetime(df['Date'])
df2 = df[df['Date'].dt.year == 2019]
```

```python
fig,ax1 = plt.subplots(figsize=(20,10))
ax1.plot(df['Date'], df['Value'])
ax1.set_ylabel('Gas Pricing,$MMBTU', size=18)
ax1.set_xlabel('Date', size=18, )


left, bottom, width, height = [0.6, 0.6, 0.25, 0.2]
ax2 = fig.add_axes([left, bottom, width, height])
ax2.plot(df2['Date'],df2['Value'])
#AX1
ax1.tick_params(labelsize=14)
ax2.tick_params('x',labelrotation=90,labelsize=14)
ax2.set_ylabel('Gas Pricing,$MMBTU', size=18)
ax2.set_xlabel('Date', size=18)
ax2.set_yticks([2.75,2.50])
ax2.tick_params(labelsize=14)
```

```python
#Mean, max and min lines line
mean  = [df['Value'].mean() for x in range(len(df['Value']))]
MIN = [df['Value'].min() for x in range(len(df['Value']))]
MAX = [df['Value'].max() for x in range(len(df['Value']))]


fig,ax3 = plt.subplots(figsize=(20,10))
ax3.plot(df['Date'],df['Value'] , c='r')
ax3.tick_params(labelsize=18)
ax3.set_yticks([5,10,15])
ax3.set_xlabel('Date', size=18)
ax3.set_ylabel('Gas Pricing, $/MMBTU', size=18)

ax3.plot(df['Date'], mean, linestyle='--', c='navy', label='MEAN')
ax3.plot(df['Date'], MIN, linestyle='--', c='salmon', label='MIN')
ax3.plot(df['Date'], MAX, linestyle='--', c='olive', label='MAX')
ax3.legend(fontsize=24)
ax3.set_title('Gas Pricing Vs. Date', fontsize=24)
```
